```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import fmin
```

```python
def banana(x): 
    """Define the Rosenbrock function
    
    Args: 
        x (numpy.ndarray): 1-D array with shape (n,) 
        
    Returns:
        Rosenbrock function
    """
    
    return 100. * (x[1] - x[0]**2)**2 + (1. - x[0])**2
```

```python
# find minimum of banana 
res = fmin(banana, [-1.2, 1], full_output=True, disp=False)
print('x:', res[0],
      '\nfval:', res[1])
```

```python
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 20
```

```python
fig = plt.figure(figsize=(16,7))

# Plot contours of banana with minimum 
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x, y)
Z = 100. * (Y - X**2)**2 + (1. - X)**2
ax = fig.add_subplot(1, 2, 1)
ax.contour(X, Y, Z, 50)
ax.plot(res[0][0], res[0][1], 'rx', ms=20, label='minimum at (1,1)')
ax.legend(loc='best')
ax.set_aspect('equal')
ax.invert_xaxis()
ax.set_xlabel(r'x')
ax.set_ylabel(r'y')

# Visualize the contour surface in 3D
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.viridis, linewidth=0.2, antialiased=False, zorder=0)
ax.plot(res[0][0], res[0][1], res[1], 'rx', ms=20, zorder=10)
ax.invert_xaxis()
ax.set_xlabel(r'x')
ax.set_ylabel(r'y')

fig.tight_layout()
plt.show()
```
