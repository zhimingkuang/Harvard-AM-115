```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp
```

```python
def poisson(lam, k):
    """Compute analytical Poisson distribution

    Args:
        lam (float): expected events per unit time (lambda)
        k (int): number of occurence

    Returns:
        float: analytical Poisson distribution
    """
    
    return lam**k / factorial(k) * exp(-lam)
```

```python
# Initialize parameters
lam = 0.5
ks = np.arange(6)

# Numerical method for Poisson distribution
trial_count = 100000
num_poisson = np.random.poisson(lam, trial_count)

# Analytical solution to Poisson distribution
ana_poisson = [poisson(lam, k) for k in ks]
```

```python
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 20
```

```python
# Visualize Poisson distribution
fig, ax = plt.subplots(1,1,figsize=(9,7))

# Plot numerical, note that `density` normalizes output
ax.hist(num_poisson, bins=ks, density=True)
# Plot analytical
ax.plot(ana_poisson, '-o', linewidth=1.5, mfc='none', ms=5.5, mew=1.5)

# Set axes labels
ax.set_xlabel(r'$k$')
ax.set_ylabel('Poisson distribution')

plt.show()
```
