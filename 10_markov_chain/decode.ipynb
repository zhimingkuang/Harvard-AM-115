```python
import numpy as np
```

```python
def compute_score(decoded_index, logM):
    score=0
    for i in range(len(decoded_index)-1):
        score+=logM[decoded_index[i+1],decoded_index[i]]
    return score
```

```python
file1=open("./Cryptography/AustenCount.txt","r")
M=np.transpose(np.loadtxt(file1))+1
logM=np.log(M)
file2=open("./Cryptography/EncryptedMessage.txt","r")
text=file2.read()
```

```python
index=np.zeros(len(text),dtype='int')
for i in range(len(text)):
    index[i]=ord(text[i])-97
    if index[i]==-65:
        index[i]=26
```

```python
map_new=np.linspace(0,26,27,dtype='int')
decoded_index=np.zeros(len(text),dtype='int')
for i in range(len(text)):
    decoded_index[i]=map_new[int(index[i])]
print(decoded_index)
```

```python
score_old=compute_score(decoded_index,logM)
for iter in range(4001):
    swap=np.random.randint(0,26,2)
    map_old=map_new.copy()
    map_new[swap[0]]=map_old[swap[1]]
    map_new[swap[1]]=map_old[swap[0]]
    for i in range(len(text)):
        decoded_index[i]=map_new[index[i]]
    score_new=compute_score(decoded_index,logM)
    if np.random.rand()>np.exp(score_new-score_old):
        map_new=map_old.copy()
    else:
        score_old=score_new
    if iter%50==0:
        decoded_text=np.zeros(len(text),dtype='str')
        for i in range(len(text)):
            if decoded_index[i]==26:
                decoded_index[i]=-65
            decoded_text[i]=chr(decoded_index[i]+97)
        out_str = ''
        for num in decoded_text:
            out_str+=num
        print(out_str)
```
