```python
# Load in data if run the notebook locally
data = np.genfromtxt('us_census_data.csv', delimiter=',', skip_header=3)
```

```python
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
```

```python
# Plot population w.r.t. year directly from data
fig, ax = plt.subplots(1,1,figsize=(9,7))
ax.plot(data[:,0], data[:,1], linewidth=1.5)
# We can alternatively set the title font style here
ax.set_title('U.S. Population with year', fontname='Times New Roman', size=24, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('U.S. Population (millions)')

plt.show()
```

```python
# Compute annual growth rate w.r.t. year with central difference method
year = data[:,0]
pop = data[:,1]
annual_growth = np.divide(pop[2:-1]-pop[0:-3], pop[1:-2]) / 20
```

```python
# Plot annual growth w.r.t. year
fig, ax = plt.subplots(1,1,figsize=(9,7))
ax.plot(year[1:-2], annual_growth, linewidth=1.5)
ax.set_title('Annual growth rate with year', fontname='Times New Roman', size=24, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('U.S. Annual Per Capita Growth Rate')

plt.show()
```

```python
# Plot annual growth w.r.t. population
fig, ax = plt.subplots(1,1,figsize=(9,7))
ax.plot(pop[1:-2], annual_growth, linewidth=1.5)
ax.set_title('Annual growth rate with population', fontname='Times New Roman', size=24, fontweight='bold')
ax.set_xlabel('U.S. Population (millions)')
ax.set_ylabel('U.S. Annual Per Capita Growth Rate')

plt.show()
```

```python
# Combine the three plots together
fig, axs = plt.subplots(1,3,figsize=(20,7))

# Plot population w.r.t. year directly from data
axs[0].plot(data[:,0], data[:,1], linewidth=1.5)
axs[0].set_title('U.S. Population with year', size=18, fontweight='bold')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('U.S. Population (millions)')

# Plot annual growth w.r.t. year
axs[1].plot(year[1:-2], annual_growth, linewidth=1.5)
axs[1].set_title('Annual growth rate with year', size=18, fontweight='bold')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('U.S. Annual Per Capita Growth Rate')

# Plot annual growth w.r.t. population
axs[2].plot(pop[1:-2], annual_growth, linewidth=1.5)
axs[2].set_title('Annual growth rate with population', size=18, fontweight='bold')
axs[2].set_xlabel('U.S. Population (millions)')
axs[2].set_ylabel('U.S. Annual Per Capita Growth Rate')

fig.tight_layout()
plt.show()
```
