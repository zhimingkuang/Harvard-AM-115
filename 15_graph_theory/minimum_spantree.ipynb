```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
import scipy
import networkx as nx
```

```python
W = [.41, .29, .51, .32, .50, .45, .38, .32, .36, .29, .21];
UGS = scipy.sparse.csr_matrix((W, ([0, 0, 1, 1, 2, 3, 3, 4, 4, 5, 5],[1, 5, 2, 4, 3, 0, 5, 2, 3, 1, 4])), shape=(6,6))
UG = nx.from_scipy_sparse_matrix(UGS)
labels = nx.get_edge_attributes(UG, 'weight')
pos = nx.planar_layout(UG)
network = nx.draw_networkx(UG, pos)
nx.draw_networkx_edge_labels(UG, pos, edge_labels=labels)
plt.show()
```

```python
ST = nx.minimum_spanning_tree(UG)
pos = nx.spiral_layout(ST)
network = nx.draw_networkx(ST, pos)
```
