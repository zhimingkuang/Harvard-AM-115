```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import sparse
import networkx as nx
```

```python
W = [.41, .99, .51, .32, .15, .45, .38, .32, .36, .29, .21]
DGS = scipy.sparse.csr_matrix((W, ([5, 0, 1, 1, 2, 3, 3, 4, 4, 5, 0],[1, 5, 2, 4, 3, 0, 5, 2, 3, 2, 4])), shape=(6,6))
DG = nx.from_scipy_sparse_array(DGS, create_using=nx.DiGraph)
labels = nx.get_edge_attributes(DG,'weight')
pos = nx.planar_layout(DG)
network = nx.draw_networkx(DG, pos)
nx.draw_networkx_edge_labels(DG, pos, edge_labels=labels)
plt.show()
```

```python
network = nx.draw_networkx(DG, pos)
nx.draw_networkx_edge_labels(DG, pos, edge_labels=labels)
path = nx.shortest_path(DG, 0, 5, weight='weight')
path_edges = zip(path, path[1:])
path_edges = set(path_edges)
nx.draw_networkx_nodes(DG, pos, nodelist=path, node_color='r')
nx.draw_networkx_edges(DG, pos, edgelist=path_edges, edge_color='r', width=2)
plt.show()
```

```python
UG = nx.DiGraph.to_undirected(DG)
labels = nx.get_edge_attributes(UG, 'weight')
pos = nx.planar_layout(UG)
network = nx.draw_networkx(UG, pos)
nx.draw_networkx_edge_labels(UG, pos, edge_labels=labels)
path = nx.shortest_path(UG, 0, 5, weight='weight')
path_edges = zip(path, path[1:])
path_edges = set(path_edges)
nx.draw_networkx_nodes(UG, pos, nodelist=path, node_color='r')
nx.draw_networkx_edges(UG, pos, edgelist=path_edges, edge_color='r', width=2)
plt.show()
```
